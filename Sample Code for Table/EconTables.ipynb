```python
#----------------------------------------------
#MAKE SURE THIS IS CHANGED!!!
observation_end = "2018-12-31"
#-------------------------------------------------

#Creates the list for Name of Release, release series name, and release series ID
release_name = ['Employment Situation', ' Unemployment Insurance Weekly Claims Report', 'Employment Situation', 
                'Manufacturers Shipments, Inventories, and Orders (M3) Survey', 'Manufacturers Shipments, Inventories, and Orders (M3) Survey',
               'G.17 Industrial Production and Capacity Utilization']
series_name = ['Unemployment Rate', 'Average Weekly Initial Claims', 'Average Weekly Earnings', 'Value of Manufacturers New Orders', 
               'Manufacturers New Orders: Nondefense Capital Goods Excluding Aircraft', 'Capacity Utilization: Manufacturing']
series_id = ['UNRATE', 'ICSA', 'CES0500000011', 'ACOGNO', 'NEWORDER', 'MCUMFN']

source = [
    'http://www.bls.gov/ces/',
    'https://www.dol.gov/ui/data.pdf',
    'http://www.bls.gov/ces/',
    'https://www.census.gov/manufacturing/m3/index.html',
    'https://www.census.gov/manufacturing/m3/index.html',
    'https://www.federalreserve.gov/releases/g17/',
]


#Creates the Ouput
values = []

for id in series_id:
    #Creates the Target URL
    data_url = 'https://api.stlouisfed.org/fred/series/observations?series_id=' + id + '&observation_end=' + observation_end + '&api_key=' + api_key + '&file_type=json' 
    response = requests.get(data_url).json()
    length = len(response['observations']) -1
    values.append(response['observations'][length]['value'])

data_df = {
    "Release" : release_name,
    "Series" : series_name,
    "Data" : values,
    "Link" : source,
    "As-Of" : observation_end,    
}

data_df = pd.DataFrame(data_df)
data_df
```
